```python
#import these laibraries
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
#insert url which website you want to scrap
urls= ['https://www.samaa.tv']
#scrape elements
for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    response.content 
```

```python
   response.content
```

```python
#replace \n with space
alldata=soup.text.replace('\n',"")
alldata
```

```python
#Removing all Digits
RemovDigit= re.sub(r'[0-9]+', '',alldata)
RemovDigit
```

```python
#Removing all Capital latters
RemoveCLatters=re.sub(r'[A-Z]+', '',RemovDigit)
RemoveCLatters
```

```python
#removing all small latters 
RemoveSLatters=re.sub(r'[a-z]+', '',RemoveCLatters)
RemoveSLatters
```

```python
RemoveSLatters.strip()
```

```python
#removing irrelivent data
fstring = RemoveSLatters.replace(u'\xa0', u' ')
fstring
```

```python
RemoveLines=fstring.replace('—',"")
AfterRemoving=RemoveLines.replace("-","")
AfterCommasRemoving=AfterRemoving.replace(",","")
AfterCommasRemoving
```

```python
#punctuation removing
AfterPuncRemove=re.sub(r'[^\w\s]', '',AfterCommasRemoving)
AfterPuncRemove
```

```python
MyString=AfterPuncRemove.split()
```

```python
#print urdu data after removing all irrelivent words
df=pd.DataFrame(MyString)
df.head
```

```python
#assign name to column
df.columns= ["اردو الفاظ"]
```

```python
#drop null value if exist 
df.dropna()
df.head(1000)
```

```python
#count total words
df.count()
```

```python
#Count reapetation of words and assign column
FinalAllWords=df.value_counts().reset_index(name='تعداد')
FinalAllWords
```
